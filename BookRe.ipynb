```python
import numpy as np
import pandas as pd
```

```python
# loading the data from the csv file to apandas dataframe
df = pd.read_csv('Books.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# Convert 'Year-Of-Publication' from object to integer
df['Year-Of-Publication'] = pd.to_numeric(df['Year-Of-Publication'], errors='coerce', downcast='integer')
```

```python
df.info()
```

```python
df = df[df['Year-Of-Publication'] >= 1990]  #taking song only after 1990
```

```python
# Set a random seed for reproducibility (optional)
np.random.seed(0)

# Number of rows to drop
n_rows_to_drop = 180000

# Randomly select 'n_rows_to_drop' rows to drop
rows_to_drop = np.random.choice(df.index, size=n_rows_to_drop, replace=False)

# Drop the selected rows from the DataFrame
df = df.drop(index=rows_to_drop)
```

```python
# resetting the index from 1 t0 ...
df = df.reset_index(drop=True)
df
```

```python
df.shape
```

```python
# selecting the relevant features for recommendation

selected_features = ["Book-Title",	"Book-Author",	"Year-Of-Publication"	]
selected_features
```

```python
# replacing the null valuess with null string

for feature in selected_features:
  df[feature] = df[feature].fillna('')
```

```python
# combining all the selected features

combined_features = (
    df['Book-Title'].astype(str) + ' ' +
    df['Book-Author'] + ' ' +
    df['Year-Of-Publication'].astype(str)
)
```

```python
print(combined_features)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# converting the data to feature vectors

vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
# getting the similarity scores using cosine similarity

similarity = cosine_similarity(feature_vectors)
```

```python
similarity.shape
```

```python
df[df['Book-Author']=="J. K. Rowling"]
```

```python
def recommendation(book_df):
    idx = df[df['Book-Title'] == book_df].index[0]
    distances = sorted(list(enumerate(similarity[idx])),reverse=True,key=lambda x:x[1])

    books = []
    for m_id in distances[1:6]:
        books.append(df.iloc[m_id[0]]['Book-Title'])

    return books
```

```python
recommendation("Quidditch Through the Ages")
```
